# **LEGENDA**<br>
**id** - ID
<br>
**battery_power** -  Całkowita energia, którą bateria może przechowywać w jednym czasie [w mAh]
<br>
**blue** -  Ma Bluetooth lub nie
<br>
**clock_speed** -  szybkość, z jaką mikroprocesor wykonuje instrukcje
<br>
**dual_sim** -  Ma obsługę dwóch kart SIM lub nie
<br>
**fc** -  Przednia kamera mega piksele<br>
**four_g** -  Ma 4G lub nie<br>
**int_memory** -  Pamięć wewnętrzna w gigabajtach<br>
**m_dep** -  Grubość komórki w cm<br>
**mobile_wt** -  Waga telefonu komórkowego<br>
**n_cores** -  Liczba rdzeni procesora<br>
**pc** -  Mega piksele kamery podstawowej<br>
**px_height** -  Wysokość rozdzielczości pikseli<br>
**px_width** -  Szerokość rozdzielczości pikseli<br>
**ram** -  Pamięć RAM w megabajtach<br>
**sc_h** -  Wysokość ekranu w cm<br>
**sc_w** -  Szerokość ekranu w cm<br>
**talk_time** -  Maksymalny czas działania telefonu podczas rozmowy<br>
**three_g** -  Ma 3G lub nie<br>
**touch_screen** -  Ma ekran dotykowy lub nie<br>
**wifi** -  Ma wifi lub nie<br>
**price_range** - Zmienna objaśniana: 0 (niska cena), 1 (średnia cena), 2 (wysoka cena), 3 (b. wysokacena).<br>

```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
mobiles=pd.read_csv("C:\\Users\\Dell\\Desktop\\python\\Egzamin\\mobile-price-classification_train.csv")
```

# Czyszczenie danych


```python
display(mobiles.head(3))
display(mobiles.tail(3))
```

```python
#sprawdzamy NAN wartosci, jesli sie pojawia
mobileste=mobiles[mobiles.isnull().any(axis=1)]
mobileste.head(2)
```

```python

#tobechanged=['blue','dual_sim','four_g','three_g','touch_screen', 'wifi']
mobiles['blue']=mobiles['blue'].transform(lambda x: "Tak" if x==1 else "Nie" )
mobiles['dual_sim']=mobiles['dual_sim'].transform(lambda x: "Tak" if x==1 else "Nie" )
mobiles['four_g']=mobiles['four_g'].transform(lambda x: "Tak" if x==1 else "Nie" )
mobiles['three_g']=mobiles['three_g'].transform(lambda x: "Tak" if x==1 else "Nie" )
mobiles['touch_screen']=mobiles['touch_screen'].transform(lambda x: "Tak" if x==1 else "Nie" )
mobiles['wifi']=mobiles['wifi'].transform(lambda x: "Tak" if x==1 else "Nie" )
def priceRange(x):
    if x== 0:
        return "1_Niska cena"
    elif x== 1:
        return "2_Średnia cena"
    elif x== 2:
        return "3_Wysoka cena"
    elif x== 3:
        return "4_B. wysoka cena"
mobiles['price_range1']=mobiles['price_range'].transform(lambda x: priceRange(x) )
```

```python
fmobiles=mobiles[["price_range1",'battery_power','clock_speed','fc','int_memory','m_dep','mobile_wt','ram']]
fmobiles.sort_values(by="price_range1",ascending=False)
ffmobiles=fmobiles.groupby("price_range1")
fig = plt.figure(figsize=(17,15))
fig.tight_layout(pad=2)
fig.suptitle('Wartości poszczególnych parametrów telefonów komórkowych \n w określonych przedziałach cenowych', fontsize=15)
ax=fig.add_subplot(2,3,1)
ax=sns.lineplot(y='battery_power',x="price_range1", data=fmobiles,palette="pastel",marker="o",markersize=7,ci=None)
ax.set_title("Całkowita energia, którą bateria \n może przechowywać w jednym czasie [w mAh] \n w poszczególnych przedziałach cenowych", fontsize=11)
ax.set_xlabel('')
ax=fig.add_subplot(2,3,2)
ax=sns.lineplot(y='clock_speed',x="price_range1", data=fmobiles,palette="pastel",marker="o",markersize=7,ci=None)
ax.set_title("Szybkość, z jaką mikroprocesor wykonuje instrukcje \n w poszczególnych przedziałach cenowych", fontsize=11)
ax.set_xlabel('')
ax=fig.add_subplot(2,3,3)
ax=sns.lineplot(y='fc',x="price_range1", data=fmobiles,palette="pastel",marker="o",markersize=7,ci=None)
ax.set_title("Przednia kamera [mln Mega Piksele] \n w poszczególnych przedziałach cenowych", fontsize=11)
ax.set_xlabel('')
ax=fig.add_subplot(2,3,4)

ax=sns.lineplot(y='int_memory',x="price_range1", data=fmobiles,palette="pastel",marker="o",markersize=7,ci=None)
ax.set_title("Pamięć wewnętrzna [w GB] \n w poszczególnych przedziałach cenowych", fontsize=11)
ax.set_xlabel('')
ax=fig.add_subplot(2,3,5)

ax=sns.lineplot(y='m_dep',x="price_range1", data=fmobiles,palette="pastel",marker="o",markersize=7,ci=None)
ax.set_title("Grubość komórki [w CM] \n w poszczególnych przedziałach cenowych", fontsize=11)
ax.set_xlabel('')
ax=fig.add_subplot(2,3,6)

ax=sns.lineplot(y='ram',x="price_range1", data=fmobiles,palette="pastel",marker="o",markersize=7,ci=None)
ax.set_title("Wielkość pamięci RAM \n w poszczególnych przedziałach cenowych", fontsize=11)
ax.set_xlabel('')
```

```python
    
fig = plt.figure(figsize=(12,20))
fig.suptitle('Zależności poszczególnych parametrów telefonów komórkowych \n w określonych przedziałach cenowych', fontsize=15)
fig.tight_layout(pad=2)
ax=fig.add_subplot(3,1,1)
ax=sns.lineplot(y='battery_power',x="clock_speed", hue="price_range1",data=fmobiles,marker="o",ci=None)
ax.legend (loc="lower right");
ax=fig.add_subplot(3,1,2)
ax=sns.lineplot(y='fc',x="int_memory", hue="price_range1",data=fmobiles,marker="o",ci=None)
ax=fig.add_subplot(3,1,3)
ax=sns.lineplot(x='m_dep',y="mobile_wt",data=fmobiles,marker="o", hue="price_range1",ci=None)
ax.legend (loc="lower right");
#fmobiles["battery_power"].describe()
    
```
